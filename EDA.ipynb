```python
import pandas as pd
import numpy as np
import csv
import base64
```

```python
def convertBoxes(num_boxes, boxes):
    return np.frombuffer(base64.b64decode(boxes), dtype=np.float32).reshape(num_boxes, 4)
def convertFeature(num_boxes, features,):
    return np.frombuffer(base64.b64decode(features), dtype=np.float32).reshape(num_boxes, 2048)
def convertLabel(num_boxes, label):
    return np.frombuffer(base64.b64decode(label), dtype=np.int64).reshape(num_boxes)
```

```python
data_path = "../data/_1_Task1/"
train = pd.read_csv(data_path + "train.tsv", sep='\t', chunksize=10, quoting=csv.QUOTE_NONE)
```

```python
for t in train:
    tmp = pd.DataFrame(t)
    break
tmp
```

```python
tmp["boxes"] = tmp.apply(lambda x: convertBoxes(x["num_boxes"], x["boxes"]), 1)
tmp["features"] = tmp.apply(lambda x: convertFeature(x["num_boxes"], x["features"]), 1)
tmp["class_labels"] = tmp.apply(lambda x: convertLabel(x["num_boxes"], x["class_labels"]), 1)


tmp
```
